NLPCC-2026-Task6-Detection/data/train_data.json 转换后得到的 NLPCC-2026-Task6-Detection/data/train_data.csv 格式如下：

```python
import json
import pandas as pd

# 输入输出文件
json_file = "NLPCC-2026-Task6-Detection/data/train_data.json"
csv_file = "NLPCC-2026-Task6-Detection/data/train_data.csv"

# 读取 JSON 文件
with open(json_file, "r", encoding="utf-8") as f:
    data = json.load(f)

# 标签映射
label_map = {
    "HWT": 0,
    "HLT": 1,
    "LGT": 2
}

# 保存转换后的数据
rows = []

for item in data:
    base_id = item["ID"]

    for key in ["HWT", "HLT", "LGT"]:
        text = item.get(key, "").strip()

        # 构造唯一ID
        new_id = f"{base_id}_{key}"

        rows.append({
            "id": new_id,
            "prompt_id": label_map[key],
            "text": text,
            "generated": label_map[key]
        })

# 转换为 DataFrame
df = pd.DataFrame(rows)

# 保存 CSV
df.to_csv(csv_file, index=False, encoding="utf-8-sig")

print(f"转换完成，共 {len(df)} 条数据")
print(df.head())
```

```python
import json
import torch
from tqdm import tqdm
from datasets import Dataset
from transformers import (
    AutoTokenizer,
    AutoModelForSequenceClassification,
    DataCollatorWithPadding
)
from torch.utils.data import DataLoader

# ======================
# Paths
# ======================

MODEL_PATH = "model-nlpcc"
TEST_FILE = "NLPCC-2026-Task6-Detection/data/testp1.json"
OUTPUT_FILE = "distilbert_submission/testp1/prediction.json"

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

# ======================
# Load model
# ======================

print("Loading model...")

tokenizer = AutoTokenizer.from_pretrained(MODEL_PATH)

model = AutoModelForSequenceClassification.from_pretrained(
    MODEL_PATH
)

model.to(DEVICE)
model.eval()

# ======================
# Load test data
# ======================

print("Loading test data...")

with open(TEST_FILE, "r", encoding="utf-8") as f:
    test_data = json.load(f)

# 保存id顺序
ids = [item["id"] for item in test_data]

dataset = Dataset.from_list(
    [{"text": item["text"]} for item in test_data]
)

# ======================
# Tokenization
# ======================

def tokenize_function(examples):
    return tokenizer(
        examples["text"],
        truncation=True,
        padding=False,
        max_length=512
    )

dataset = dataset.map(
    tokenize_function,
    batched=True
)

dataset.set_format(
    type="torch",
    columns=["input_ids", "attention_mask"]
)

data_collator = DataCollatorWithPadding(
    tokenizer=tokenizer
)

dataloader = DataLoader(
    dataset,
    batch_size=32,
    shuffle=False,
    collate_fn=data_collator
)

# ======================
# Prediction
# ======================

print("Predicting...")

predictions = []

with torch.no_grad():
    for batch in tqdm(dataloader):

        batch = {
            k: v.to(DEVICE)
            for k, v in batch.items()
        }

        outputs = model(**batch)

        preds = torch.argmax(
            outputs.logits,
            dim=-1
        )

        predictions.extend(
            preds.cpu().numpy().tolist()
        )

# ======================
# Save submission file
# ======================

submission = []

for sample_id, pred in zip(ids, predictions):
    submission.append({
        "id": sample_id,
        "label": int(pred)
    })

with open(
    OUTPUT_FILE,
    "w",
    encoding="utf-8"
) as f:
    json.dump(
        submission,
        f,
        ensure_ascii=False,
        indent=2
    )

print(f"Saved: {OUTPUT_FILE}")
print(f"Total predictions: {len(submission)}")
```

```python
import json
import torch
from tqdm import tqdm
from datasets import Dataset
from transformers import (
    AutoTokenizer,
    AutoModelForSequenceClassification,
    DataCollatorWithPadding
)
from torch.utils.data import DataLoader

# ======================
# Paths
# ======================

MODEL_PATH = "model-nlpcc"
TEST_FILE = "NLPCC-2026-Task6-Detection/data/testp2.json"
OUTPUT_FILE = "distilbert_submission/testp2/prediction.json"

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

# ======================
# Load model
# ======================

print("Loading model...")

tokenizer = AutoTokenizer.from_pretrained(MODEL_PATH)

model = AutoModelForSequenceClassification.from_pretrained(
    MODEL_PATH
)

model.to(DEVICE)
model.eval()

# ======================
# Load test data
# ======================

print("Loading test data...")

with open(TEST_FILE, "r", encoding="utf-8") as f:
    test_data = json.load(f)

# 保存id顺序
ids = [item["id"] for item in test_data]

dataset = Dataset.from_list(
    [{"text": item["text"]} for item in test_data]
)

# ======================
# Tokenization
# ======================

def tokenize_function(examples):
    return tokenizer(
        examples["text"],
        truncation=True,
        padding=False,
        max_length=512
    )

dataset = dataset.map(
    tokenize_function,
    batched=True
)

dataset.set_format(
    type="torch",
    columns=["input_ids", "attention_mask"]
)

data_collator = DataCollatorWithPadding(
    tokenizer=tokenizer
)

dataloader = DataLoader(
    dataset,
    batch_size=32,
    shuffle=False,
    collate_fn=data_collator
)

# ======================
# Prediction
# ======================

print("Predicting...")

predictions = []

with torch.no_grad():
    for batch in tqdm(dataloader):

        batch = {
            k: v.to(DEVICE)
            for k, v in batch.items()
        }

        outputs = model(**batch)

        preds = torch.argmax(
            outputs.logits,
            dim=-1
        )

        predictions.extend(
            preds.cpu().numpy().tolist()
        )

# ======================
# Save submission file
# ======================

submission = []

for sample_id, pred in zip(ids, predictions):
    submission.append({
        "id": sample_id,
        "label": int(pred)
    })

with open(
    OUTPUT_FILE,
    "w",
    encoding="utf-8"
) as f:
    json.dump(
        submission,
        f,
        ensure_ascii=False,
        indent=2
    )

print(f"Saved: {OUTPUT_FILE}")
print(f"Total predictions: {len(submission)}")
```
